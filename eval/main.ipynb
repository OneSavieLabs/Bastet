```python
import requests
import pandas as pd
from pandas import read_csv
import sklearn.metrics
from pprint import pprint

url = 'http://scat-n8n.dev.onesavie.com/webhook/0f3a3dc2-9cb2-4454-86b1-bb95ef26a901'

file_path = 'dataset.csv'
dataset = read_csv(file_path)
y_pred = []
y_true = []
for index, row in dataset.iterrows():
    if (pd.isna(row['file_name']) or pd.isna(row['answer'])):
        break
    with open("./source_code/"+row['file_name']) as f:
        file = f.read()
        data = {
            "prompt": file
        }
        response = requests.post(url, json=data)
        pprint(response.status_code)
        try:
            json_data = response.json()
            y_pred.append(1 if json_data["output"] else 0)
            y_true.append(int(row['answer']))
        except:
            print(response.text)
            continue
tn, fp, fn, tp = sklearn.metrics.confusion_matrix(y_pred=y_pred, y_true=y_true, labels=[0,1]).ravel()
print("True Positive: ", tp)
print("True Negative: ", tn)
print("False Positive: ", fp)
print("False Negative: ", fn)
```
